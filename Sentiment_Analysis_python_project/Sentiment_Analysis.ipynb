```python
# project- sentimental_Analysis uing wordcloud

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from textblob import TextBlob   #TextBlob is used for Natural Language Processing (NLP).Sentiment Analysis,Spell correction, Text classification
from wordcloud import WordCloud  #WordCloud is used to visualize text data.
pd.set_option('display.max_colwidth', None)

#load dataset
data= pd.read_csv("Sentiment_Analysis.csv")
data.head(20)
```

```python

def get_sentiment(text):
    analysis = TextBlob(text)

    if analysis.sentiment.polarity > 0:    #sentence polarity as Nutural (0) ,Postive (1), negative (-1)
        return "Positive"


    elif analysis.sentiment.polarity < 0:
       return "Negative"

    else:
       return "Nutral"
    
    #Apply sentiment Analysis as a column in dataset to show sentance sentiment

data["Sentiment"] = data["Sentences"].apply(get_sentiment)
data.head(20)
```

```python
# to download dataset with text sentiment
#data. to_excel("Manu_kumar_sentiment3.xlsx",index = False)
```

```python
#filtring negative sentiment

negative_reviews = data[data["Sentiment"] == "Negative"]
negative_reviews.head(10)
```

```python
#combine all negative sentances in one string
negative_merge = " ".join(negative_reviews["Sentences"])
print(negative_merge)
```

```python
#create the wordcloud of negative reviews

visual_of_wordcloud = WordCloud(width = 800, height = 400,background_color ="black").generate(negative_merge)
plt.figure(figsize=(10,5))
plt.title("Most common words in negative reviews",fontname="calibri",color="red",fontstyle="italic")
plt.imshow(visual_of_wordcloud,interpolation ="bilinear")
plt.axis("off")
plt.show()
```
