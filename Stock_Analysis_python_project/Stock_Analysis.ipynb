```python
# importing inbuilt libraries 

import numpy as np                 #numrical computiation
import pandas as pd                #Data Manupulation and analysis
import matplotlib.pyplot as plt    #Data visualization
import seaborn as sns              #Advanced data visualization
import yfinance as yf              #Fetch stock market data

sns.set_style("darkgrid")

print("============stock Analysis=========")

#using while loop here because i don't know how many time will check for stcoks price or trend
while True:
       #storing input data in symbol variable and using upper here to convert stock symbol name in upper latter
    symbol= input("Enter the stock symble (or 'Q' to quit the program):").upper()

    if symbol == "Q":
        print("Exiting the program")
        break

    try:
            #Ticker func will go to Yahoo_finance and look for symbol and store in stock variable
            stock= yf.Ticker(symbol)
            info= stock.info    #info function will take lots of info from stock variable and store in info
            #taking the below mentioned details from info variable for that using .get function
            name= info.get("longName","NA")
            price= info.get("currentPrice")
            prev_close= info.get("previousClose")
            currency= info.get("currency","USD")
            print("\n-----------------------------------------")
            #printing below details of stocks
            print(f"company:{name}")
            print(f"currentprice:{price}")

            if price and prev_close:
                change = price - prev_close

                percent=(change/prev_close)*100

                print(f" change:{change:+.2f} ({percent:+.2f}%)")

                print("\n-----------------------------------------")

                #downloading historical data from yahoo finance and starting date and ending date as mentiond on the interval of one day

                df= yf.download(symbol,start ="2023-01-01",end="2026-01-01", interval="1d")

                #creating new column inside df name as 7_dayMovingAverage and using rolling func calculation moving average
                #creating new column inside df name as 30_dayMovingAverage and using rolling func calculation moving average

                df["7_dayMovingAverage"] =df["Close"].rolling(window=7).mean()
                df["30_dayMovingAverage"] =df["Close"].rolling(window=30).mean()

                # ploting grap for stock using line chart

                plt.figure(figsize=(10,5))
                plt.plot(df["Close"])
                plt.plot(df["7_dayMovingAverage"])
                plt.plot(df["30_dayMovingAverage"])
                plt.title(f"{symbol} stock price analysis")

                plt.show()

    except Exception as e:
            print(e)


```
